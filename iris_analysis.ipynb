```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
iris=pd.read_csv('D:/dataset/Iris.csv')
iris.head()
```

```python
iris.drop('Id',axis=1,inplace=True)
print(iris.columns)
print(iris["Species"].value_counts())
iris.head()
```

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(iris,hue='Species',size=5)
plt.show()
```

```python
versicolor=iris.loc[iris['Species']=='Iris-versicolor']
setosa=iris.loc[iris['Species']=="Iris-setosa"]
virginica=iris.loc[iris['Species']=='Iris-virginica']
```

```python
plt.plot(versicolor['PetalLengthCm'],np.zeros(len(versicolor)),'o')
plt.plot(setosa['PetalLengthCm'],np.zeros(len(setosa)),'o')
plt.plot(virginica['PetalLengthCm'],np.zeros(len(virginica)),'o')
plt.grid()
plt.show()
```

```python
sns.FacetGrid(iris,hue='Species',size=6).map(sns.distplot,'PetalLengthCm').add_legend()
plt.show()
```

```python
sns.FacetGrid(iris,hue='Species',size=6).map(sns.distplot,'SepalLengthCm').add_legend()
```

```python
sns.FacetGrid(iris,size=6,hue='Species').map(sns.distplot,'SepalWidthCm').add_legend()
```

```python
sns.FacetGrid(iris,hue='Species',size=6).map(sns.distplot,'PetalWidthCm').add_legend()
```

```python
counts,bin_edges=np.histogram(setosa['PetalLengthCm'],bins=10,density=True)
print(counts,"\n",bin_edges)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,color='g')
plt.plot(bin_edges[1:],cdf,color='r')
```

```python
#ploting the pdf and cdf for each flower
counts,bin_edges=np.histogram(setosa['PetalLengthCm'],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#pdf and cdf for versicolor
counts,bin_edges=np.histogram(versicolor['PetalLengthCm'],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#pdf and cdf for verginica
counts,bin_edges=np.histogram(virginica['PetalLengthCm'],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

```python
#mean
import numpy as np
print("means :")
print(np.mean(setosa['PetalLengthCm']))
print(np.mean(versicolor['PetalLengthCm']))
print(np.mean(virginica['PetalLengthCm']))
#mean with outlier
print("means with outlier :",end=" ")
print(np.mean(np.append(setosa['PetalLengthCm'],50)))
```

```python
#median of Petallength
print("medians :")
print(np.median(setosa['PetalLengthCm']))
print(np.median(versicolor['PetalLengthCm']))
print(np.median(virginica['PetalLengthCm']))
#median with outlier
print("median with outlier : ",end=" ")
print(np.median(np.append(setosa['PetalLengthCm'],50)))
```

```python
#quantiles
import numpy as np
print("quantiles : ")
print(np.percentile(setosa['PetalLengthCm'],np.arange(0,100,25)))
print(np.percentile(versicolor['PetalLengthCm'],np.arange(0,100,25)))
print(np.percentile(virginica['PetalLengthCm'],np.arange(0,100,25)))

print('90% quantiles :')
print(np.percentile(setosa['PetalLengthCm'],90))
print(np.percentile(versicolor['PetalLengthCm'],90))
print(np.percentile(virginica['PetalLengthCm'],90))

print('Median Absolute Deviation : ')
from statsmodels import robust
print(robust.mad(setosa['PetalLengthCm']))
print(robust.mad(versicolor['PetalLengthCm']))
print(robust.mad(virginica['PetalLengthCm']))
```
